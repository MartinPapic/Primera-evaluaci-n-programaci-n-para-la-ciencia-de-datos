En esta libreta de Jupyter, realizaremos un diagnóstico inicial e inspección de los cuatro datasets provistos:
* `asistencia.csv`
* `calificaciones.csv`
* `estudiantes.csv`
* `inscripciones.csv`

El objetivo será identificar la calidad de los datos, detectar valores nulos (~5%), duplicados (~3%), outliers y formatos inconsistentes, como paso previo a nuestro pipeline de limpieza en Kedro.


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from IPython.display import display

# Configuración visual
sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

raw_data_dir = '../data/01_raw/'
```

```
# Carga de datasets
df_asistencia = pd.read_csv(os.path.join(raw_data_dir, 'asistencia.csv'))
df_calificaciones = pd.read_csv(os.path.join(raw_data_dir, 'calificaciones.csv'))
df_estudiantes = pd.read_csv(os.path.join(raw_data_dir, 'estudiantes.csv'))
df_inscripciones = pd.read_csv(os.path.join(raw_data_dir, 'inscripciones.csv'))

datasets = {
    'Asistencia': df_asistencia,
    'Calificaciones': df_calificaciones,
    'Estudiantes': df_estudiantes,
    'Inscripciones': df_inscripciones
}

print("Datasets cargados.")
```

```
def analyze_dataset(name, df):
    print(f"\n{'='*50}")
    print(f"Dataset: {name}")
    print(f"{'='*50}")
    print(f"Shape: {df.shape}")
    
    print("\n1. Tipos de Datos y Valores No Nulos:")
    df.info()
    
    print("\n2. Valores Nulos:")
    nulls = df.isnull().sum()
    null_percent = (nulls / len(df)) * 100
    display(pd.DataFrame({'Nulos': nulls, '%': null_percent.round(2)})[nulls > 0])
    
    print(f"\n3. Filas Duplicadas: {df.duplicated().sum()}")
    
    if len(df.select_dtypes(include=[np.number]).columns) > 0:
        print("\n4. Estadísticas Descriptivas (Numéricas):")
        display(df.describe())
        
    print("\n5. Muestra de Datos:")
    display(df.head(3))
```

```
for name, df in datasets.items():
    analyze_dataset(name, df)
```

```
# Análisis Visual de Outliers en Calificaciones (Ejemplo)
# Cambia 'nota' o el nombre correcto después de ver el describe
try:
    cols_numericas = df_calificaciones.select_dtypes(include=[np.number]).columns
    if len(cols_numericas) > 0:
        plt.figure(figsize=(8,4))
        sns.boxplot(x=df_calificaciones[cols_numericas[0]].dropna())
        plt.title(f'Distribución y Outliers en {cols_numericas[0]}')
        plt.show()
except Exception as e:
    print(e)
```
